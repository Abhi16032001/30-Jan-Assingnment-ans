```python
Answer 1.

percentage = float(input("Enter your percentage: "))

if percentage > 90:

grade = 'A'

elif percentage > 80 and percentage <= 90:

grade = 'B'

elif percentage >= 60 and percentage <= 80:

grade = 'C'

else:

grade = 'D'

print(f"Your grade is {grade}")

```

```python
##Answer 6.

# Print a square pattern

n = 5
i = 1
while i <= n:
    j = 1
    while j <= n:
        print("* ", end="")
        j += 1
    print()
    i += 1
    
* * * * * 
* * * * * 
* * * * * 
* * * * * 
* * * * * 
```
